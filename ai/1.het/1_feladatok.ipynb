**8. Készítsen egy függvényt, amely megszámolja, hogy a paramétereként megadott szövegben hányszor fordul elő a `'kutya'` szó. Vegye figyelembe azt is, ha a `'kutya'` nagybetűvel írva fordul elő a szövegben! (ragokkal nem kell foglalkozni).**

**Például: `'ez a kutya gyorsabban fut mint a másik KUTYA'` esetén a visszatérési érték: `2`.**

````python
text = "ez a kutya gyorsabban fut mint a másik KUTYA"
def countKutya(text):
    count = 0
    for word in text:
        if word == "kutya" or word == "KUTYA":
            count += 1
    return count
````

````python
countKutya(text)
````

### Plusz kérdés
**Kicsit gyorsan vezet és megállítja a rendőr. Írjon függvényt, amely meghatározza, hogy kell-e és ha igen, milyen büntetést kell fizetni. A lehetséges esetek: `'Nincs büntetés'`, `'Kis büntetés'`, `'Nagy büntetés'`.**

- Ha a sebesség 60 vagy annál kisebb, `'Nincs büntetés'`. 
- Ha a sebesség 60-nál nagyobb, de 80 vagy annál kisebb, `'Kis büntetés'`. 
- Ha a sebesség 80-nál nagyobb, akkor `'Nagy büntetés'`.
- Kivéve, ha éppen születésnapja van (ezt egy logikai értékkel adja meg a függvény paramétereként) - ekkor a sebessége öttel nagyobb lehet mindegyik esetben. 
  
**Például:**

- `mi_lesz(81,True)` esetén a visszatérési érték: `'Kis büntetés'`
   
- `mi_lesz(81,False)` esetén a visszatérési érték: `'Nagy büntetés'`

````python
def mi_lesz(sebesseg, szulinap):
    if szulinap:
        sebesseg -= 5
    if sebesseg <= 60:
        return "nincs büntetés"
    if 60 <= sebesseg <= 80:
        return "kis büntetés"
    if sebesseg >= 80:
        return "nagy büntetés"
````

````python
mi_lesz(84,True)
````
